```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

# Implement Mean Absolute Error

```python
#using flat python code
def mae(y_true , y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        total_error += abs(yt - yp)
        mean_absolute_error = total_error / len(y_true)
    return mean_absolute_error
```

```python
mae(y_true , y_predicted)
```

# Implement same thing using numpy in much easier way

```python
##using numpy
np.mean(np.abs(y_true - y_predicted))
```

```python
def mae_np(y_true , y_predicted):
    mae = np.mean(np.abs(y_true - y_predicted))
    return mae
```

```python
mae_np(y_true , y_predicted)
```

# Implement Log Loss or Binary Cross Entropy

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
#for log of predicted value for log(0) its inf we have convert values near to 0 or near to 1 
```

```python
epsilon = 1e-15
np.log(epsilon)
```

```python
y_predicted_new = [max(i,epsilon)  for i in y_predicted] #zero values from y_predicted are near to zero now
y_predicted_new = [min(i,1-epsilon)  for i in y_predicted_new] #one values from y_predicted are near to one now

#convert list to numpy array
y_predicted_new = np.array(y_predicted_new)
y_predicted_new
```

```python
log_loss = -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
log_loss
```

```python
def logloss(y_true , y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon)  for i in y_predicted] #zero values from y_predicted are near to zero now
    y_predicted_new = [min(i,1-epsilon)  for i in y_predicted_new] #one values from y_predicted are near to one now

    #convert list to numpy array
    y_predicted_new = np.array(y_predicted_new)

    #apply loglos fun
    log_loss = -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
    return log_loss

logloss(y_true , y_predicted)
```

Without using numpy (i.e. using plain python)

With the use of numpy

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
#Without using numpy (i.e. using plain python)

def mean_squered_error(y_true, y_predicted):
    total_squred_sum = 0
    for yt, yp in zip(y_true, y_predicted):
        total_squred_sum += (yt - yp)**2
        mse = total_squred_sum / len(y_true)
    print("Total squred error: ",total_squred_sum)
    print("Mean squred error : ",mse)
```

```python
mean_squered_error(y_true, y_predicted)
```
